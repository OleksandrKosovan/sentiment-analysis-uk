### Функції

```python
SYSTEM_PATH = "/Users/oleksandrkosovan/Documents/GitHub/"
LEM_PATH = "nlp_uk/src/main/groovy/org/nlp_uk/tools/LemmatizeText.groovy"

def lemmatization(
    in_data_path, 
    in_file_name, 
    out_data_path, 
    out_file_name, 
    system_path=SYSTEM_PATH, 
    lem_path=LEM_PATH
):
    in_path = os.path.join(in_data_path, in_file_name)
    out_path = os.path.join(out_data_path, out_file_name)
    
    lem_command = "groovy " + os.path.join(system_path, lem_path) + " -i " + in_path + " -o " + out_path
    os.system(lem_command)
```

```python
def text_preparation(
    in_data_path,
    in_file_name,
    out_data_path,
    out_file_name,
):
    with open(os.path.join(in_data_path, in_file_name), 'r') as file:
        text = file.read()
        file.close()
    with open(os.path.join(out_data_path, out_file_name),"w") as out_put:
        out_put.write(text)
        out_put.close()
    
    lemmatization(
        in_data_path,
        in_file_name,
        out_data_path,
        out_file_name
    )
```

```python
# test

text_preparation(
    in_data_path='04-test-files',
    in_file_name='input.txt',
    out_data_path='04-test-files',
    out_file_name='output.txt',
)
```

### Підготовка тексту

```python
from joblib import Parallel, delayed
import multiprocessing
```

```python
num_cores = multiprocessing.cpu_count()
```

```python
def neg(file_name):
    IN_DATA_PATH = '01-reviews-data/negative/'
    OUT_DATA_PATH = '02-lem-data/negative/'
    try:
        text_preparation(
            in_data_path=IN_DATA_PATH,
            in_file_name=file_name,
            out_data_path=OUT_DATA_PATH,
            out_file_name=file_name
        )
    except:
        print('some error')
```

```python
IN_DATA_PATH = '01-reviews-data/negative/'
files_list = os.listdir(IN_DATA_PATH)

Parallel(
    n_jobs=num_cores, 
    verbose=50
)(delayed(neg)(file_name)for file_name in files_list)
```

```python
def poss(file_name):
    IN_DATA_PATH = '01-reviews-data/positive/'
    OUT_DATA_PATH = '02-lem-data/positive/'
    try:
        text_preparation(
            in_data_path=IN_DATA_PATH,
            in_file_name=file_name,
            out_data_path=OUT_DATA_PATH,
            out_file_name=file_name,
        )
    except:
        print('some error')
```

```python
IN_DATA_PATH = '01-reviews-data/positive/'
files_list = os.listdir(IN_DATA_PATH)

Parallel(
    n_jobs=num_cores, 
    verbose=50
)(delayed(poss)(i)for i in files_list)
```
